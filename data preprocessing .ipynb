### Dummy values for proto and conn_state:

```python
combined = pd.get_dummies(combined, columns=['proto','conn_state'])
```

```python
combined
```

```python
for col in combined.columns:
    print(col,end=", ")
```

```python
combined = combined.drop(['ts','uid','id.orig_h','id.orig_p','id.resp_h','id.resp_p',
                               'service','local_orig','local_resp','history','tunnel_parents'],axis=1)
```

```python
for col in combined.columns:
    print(col,end=", ")
```

```python
combined.interpolate(method ='linear', limit_direction ='forward', inplace=True)
```
